```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
plt.rcParams['figure.figsize'] = [20,12]
```

```python
df = pd.read_csv('UK_case_info.csv')
df
```

```python
def date_month(date):
    date=date.split("-")
    if date[1]=="01":
        return "January"+date[0]
    if date[1]=="02":
        return "February"+date[0]
    if date[1]=="03":
        return "March"+date[0]
    if date[1]=="04":
        return "April"+date[0]
    if date[1]=="05":
        return "May"+date[0]
    if date[1]=="06":
        return "June"+date[0]
    if date[1]=="07":
        return "July"+date[0]
    if date[1]=="08":
        return "August"+date[0]
    if date[1]=="09":
        return "September"+date[0]
    if date[1]=="10":
        return "October"+date[0]
    if date[1]=="11":
        return "November"+date[0]
    if date[1]=="12":
        return "December"+date[0]
```

```python
month = []
for i in df['date']:
    month.append(date_month(i))
df.insert(9,"month",month,True)
new_df = df.copy()
new_df
```

```python
def variant(date):
    date = date.split('-')
    if date[0] == '2020':
        return 'alpha'
    elif date[0] == '2021':
        if (date[1] == '01' or date[1] == '02' or date[1] == '03'):
            return 'alpha'
        elif date[1] == '12':
            return 'omicron'
        else:
            return 'delta'
    elif date[0] == '2022':
        return 'omicron'
```

```python
variants = []
for i in df['date']:
    variants.append(variant(i))
new_df.insert(10,"variant",variants,True)
new_df
```

```python
new_df['rise in cases']= ""
new_df['rise in cases'] = new_df['active_cases'].diff()
dummy= new_df.copy()
dummy.dropna(subset=['rise in cases'],inplace=True)
dummy
```

# ALPHA VARIANT

```python
def bar_graph(categories,cat_val,xlabel,ylabel,title=""):
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.bar(categories,cat_val,width=0.8)
    plt.title(title)
    plt.show()
```

```python
new2_df = dummy.groupby('variant')
alpha = new2_df.get_group('alpha')
alpha.dropna(subset=['active_cases','cases_new'],inplace=True)
alpha.sort_values(by='date',inplace=True)
alpha.reset_index(inplace=True)
alpha
```

```python
alpha_increase= alpha[alpha['rise in cases']>0]
alpha_decrease = alpha[alpha['rise in cases']<0]
alpha_increase
```

```python
start_date='2020-09-01'
end_date='2020-09-30'
sept = (alpha_increase['date'] > start_date) & (alpha_increase['date'] <= end_date)
sept_impact = alpha_increase.loc[sept]
sept_impact
```

```python
per=[]
for i in sept_impact.index:
    per.append(sept_impact['cases_new'][i]/sept_impact['tests_today'][i])
per
```

```python
alpha_decrease
```

```python
alpha_increase_stats= alpha_increase.groupby('month').sum()
alpha_increase_stats
```

```python
alpha_decrease_stats= alpha_decrease.groupby('month').sum()
alpha_decrease_stats
```

```python
bar_graph(alpha_increase_stats.index.values,alpha_increase_stats["rise in cases"],"Months","Number of cases","Average rise in Cases perday")
```

DESCRIPTION: The above graph shows the rise in cases in Uttarakhand during 1st Wave

```python
alpha.plot(x="date",y="active_cases",title="active cases in Uttarakhand during 1st Wave ")
```

# DELTA VARIANT

```python
new2_df = dummy.groupby('variant')
delta = new2_df.get_group('delta')
delta.dropna(subset=['active_cases','cases_new'],inplace=True)
delta.sort_values(by='date',inplace=True)
delta.reset_index(inplace=True)
delta
```

```python
delta.plot(x="date",y="active_cases",title="active cases in Uttarakhand during 2nd Wave ")
```

```python
result = pd.concat([alpha, delta], axis=1)
result.reset_index(inplace=True)
result
```

```python
result.plot(x="level_0",y=['active_cases'])
plt.xlabel('No of Days')
plt.ylabel('active_cases')
```
