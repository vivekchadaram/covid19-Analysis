```python

def get_monthday(date):
    date=date.split("-")
    if date[1]=="01":
        return "January"+date[0]
    if date[1]=="02":
        return "February"+date[0]
    if date[1]=="03":
        return "March"+date[0]
    if date[1]=="04":
        return "April"+date[0]
    if date[1]=="05":
        return "May"+date[0]
    if date[1]=="06":
        return "June"+date[0]
    if date[1]=="07":
        return "July"+date[0]
    if date[1]=="08":
        return "August"+date[0]
    if date[1]=="09":
        return "September"+date[0]
    if date[1]=="10":
        return "October"+date[0]
    if date[1]=="11":
        return "November"+date[0]
    if date[1]=="12":
        return "December"+date[0]
def get_month(date):
    date=date.split("-")
    if date[1]=="01":
        return "January"+date[2]
    if date[1]=="02":
        return "February"+date[2]
    if date[1]=="03":
        return "March"+date[2]
    if date[1]=="04":
        return "April"+date[2]
    if date[1]=="05":
        return "May"+date[2]
    if date[1]=="06":
        return "June"+date[2]
    if date[1]=="07":
        return "July"+date[2]
    if date[1]=="08":
        return "August"+date[2]
    if date[1]=="09":
        return "September"+date[2]
    if date[1]=="10":
        return "October"+date[2]
    if date[1]=="11":
        return "November"+date[2]
    if date[1]=="12":
        return "December"+date[2]
    
def get_year(date):
    date=date.split("-")
    return date[0]
def Histogram(x,y,xlabel,ylabel,title):
    x.plot(kind="hist",
        alpha=0.7,
        #         bins=[1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,15000],
        title=title,
        figsize=(12,8),
        color=["red"],rwidth=0.8
       )
    y.plot(kind="hist",
        alpha=0.7,
        #         bins=[1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,15000],
        title=title,
        figsize=(12,8),
        color=["blue"],rwidth=0.8
       )
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
def bar_graph(categories,cat_val,xlabel,ylabel,title="",color="blue"):
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.bar(categories,cat_val,width=0.8,color=color)
    plt.title(title)
    plt.show()
# This is the code of linear_regression model also.........           
def linear_correlation(x,y):
    meanx=sum(x.values)/len(x)
    meany=sum(y.values)/len(y)
    if len(x.values)==len(y.values):
        num,den=0,0
        #numerator calculation
        a=0
        for i in range(len(x)):
            a+=x.values[i]*y.values[i]
            #         print(a)
        num=a*len(x)-(sum(x.values)*sum(y.values))
    
        #denominator calculation
        ax,ay=0,0
        for i in range(len(x)):
                ax+=x.values[i]**2
                ay+=y.values[i]**2 
        p1=(len(x)*ax-(sum(x.values)**2))**0.5
        p2=(len(y)*ay-(sum(y.values)**2))**0.5
        den=p1*p2
        #print(num,den)
        return num/den
    else:
        raise Exception("Invalid Dimension")
import datetime
import calendar
 
def get_day(date):
    day, month, year = (int(i) for i in date.split('-'))   
    dayNumber = calendar.weekday(year, month, day)
    days =["Monday", "Tuesday", "Wednesday", "Thursday",
                         "Friday", "Saturday", "Sunday"]
    return (days[dayNumber])

def set_index(x):
    if x=="June2020":
        return -2
    if x=="July2020":
        return -1
    if x=="August2020":
        return 0
    if x=="September2020":
        return 1
    if x=="October2020":
        return 2
    if x=="November2020":
        return 3
    if x=="December2020":
        return 4
    if x=="January2021":
        return 5
    if x=="February2021":
        return 6
    if x=="March2021":
        return 7
    if x=="April2021":
        return 8
    if x=="May2021":
        return 9
    if x=="June2021":
        return 10
    if x=="July2021":
        return 11
    if x=="August2021":
        return 12
    if x=="September2021":
        return 13
    if x=="October2021":
        return 14
    if x=="November2021":
        return 15
    if x=="December2021":
        return 16
    if x=="January2022":
        return 17
def piechart(categories,cat_val,title=""):
    labels=categories
    values=cat_val
    #     explode = (0,0.1)
    plt.pie(values,labels=values,counterclock=False,autopct="%.2f", shadow=True)
    plt.title(title)
    plt.legend(labels,loc=3)
    plt.rcParams['figure.figsize'] = [10,50]
    plt.show()
def rearrange_date(x):
    arr=x.split("-")
    s="-"
    s1=s.join(arr[::-1])
    return s1
# vacc_month_sum[["vax_total_24h","vax_first_dose_24h","vax_sec_dose_24h"]].plot(kind="bar")
def plot_bar_graphs(data,values,xlabel="",ylabel="",title=""):
    data[values].plot(kind="bar",title="",xlabel=xlabel,ylabel=ylabel)
def variant(date):
    date = date.split('-')
    if date[0] == '2020':
        return 'alpha'
    elif date[0] == '2021':
        if (date[1] == '01' or date[1] == '02' or date[1] == '03'):
            return 'alpha'
        elif date[1] == '12':
            return 'omicron'
        else:
            return 'delta'
    elif date[0] == '2022':
        return 'omicron'
def TPR(total_tests,positive):
    return (positive/total_tests)*100
def comorb(s):
    if s=="no comorbidity":
        return "no comorbidity"
    else:
        return "comorbidity"
def cato(s):
    if "Government" in s or "Govt" in s or "govt" in s or "government" in s:
        return "Government facility"
    if "Private" in s :
        return "Private facility"
    if "college" in s:
        return "Medical college"
    else:
        return "Others"
def waves(date):
    date = date.split('-')
    if date[0] == '2020':
        return 'first'
    elif date[0] == '2021':
        if (date[1] == '01' or date[1] == '02' or date[1] == '03'):
            return 'first'
        else:
            return 'second'
    elif date[0] == '2022':
        return 'third'
```

```python
district_cases = pd.read_csv("MH_district_caseinfo.csv")
district_cases.sort_values(by='date',ascending=False,inplace=True)
district_refined = district_cases.head(37)
# district_refined
district_total = district_refined[district_refined['district_name'] == 'total']
# district_total
```

```python
d = district_refined[district_refined['district_name'] == 'total'].index
d1= district_refined[district_refined['district_name'] == 'other states/country'].index
district_refined.drop(d1,inplace=True)
district_refined.drop(d,inplace=True)
# district_refined
```

```python
district_refined['cases_per_total'] = ""
district_refined['cases_per_total'] = district_refined['cases_total']/district_total.iloc[0]['cases_total']
district_refined['deaths_per_total'] = ""
district_refined['deaths_per_total'] = district_refined['deaths_total']/district_total.iloc[0]['deaths_total']
# district_refined
```

```python
wave = []
for i in district_cases['date']:
    wave.append(waves(i))
district_cases.insert(7,"waves",wave,True)
```

# First wave and it's potential causes:

```python
district_cases_dummy= district_cases.groupby('waves')
first_district_cases = district_cases_dummy.get_group('first')
first_district_cases.sort_values(by='date',ascending=False,inplace=True)
d = first_district_cases[first_district_cases['district_name'] == 'total'].index
d1= first_district_cases[first_district_cases['district_name'] == 'other states/country'].index
first_district_cases.drop(d1,inplace=True)
first_district_cases.drop(d,inplace=True)

first_district_cases['per_of_deaths'] = ""
first_district_cases['per_of_deaths'] = (first_district_cases['deaths_total']/first_district_cases['cases_total'])*100
# first_district_cases
```

```python
first_district_cases_mean = first_district_cases.groupby('district_name').mean()
# first_district_cases_mean
```

```python
plt.figure(figsize=(25,35))
plt.barh(first_district_cases_mean.index.values, first_district_cases_mean['per_of_deaths'])
plt.title("Percentage of deaths in each district")
```

# Second wave and it's potential causes:

```python
second_district_cases = district_cases_dummy.get_group('second')
second_district_cases.sort_values(by='date',ascending=False,inplace=True)
d = second_district_cases[second_district_cases['district_name'] == 'total'].index
d1= second_district_cases[second_district_cases['district_name'] == 'other states/country'].index
second_district_cases.drop(d1,inplace=True)
second_district_cases.drop(d,inplace=True)

second_district_cases['per_of_deaths'] = ""
second_district_cases['per_of_deaths'] = (second_district_cases['deaths_total']/second_district_cases['cases_total'])*100
# second_district_cases
```

```python
second_district_cases_mean = second_district_cases.groupby('district_name').mean()
# second_district_cases_mean
```

```python
plt.figure(figsize=(25,35))
plt.barh(second_district_cases_mean.index.values, second_district_cases_mean['per_of_deaths'])
plt.title("Percentage of deaths in each district during 2nd Wave")
```

# Third wave and it's potential causes:

```python
third_district_cases = district_cases_dummy.get_group('second')
third_district_cases.sort_values(by='date',ascending=False,inplace=True)
d = third_district_cases[third_district_cases['district_name'] == 'total'].index
d1= third_district_cases[third_district_cases['district_name'] == 'other states/country'].index
third_district_cases.drop(d1,inplace=True)
third_district_cases.drop(d,inplace=True)

third_district_cases['per_of_deaths'] = ""
third_district_cases['per_of_deaths'] = (third_district_cases['deaths_total']/third_district_cases['cases_total'])*100
# third_district_cases.drop("district_name")
```

```python
third_district_cases_mean = third_district_cases.groupby('district_name').mean()
# third_district_cases_mean
```

```python
plt.figure(figsize=(25,35))
plt.barh(third_district_cases_mean.index.values, third_district_cases_mean['per_of_deaths'])
plt.title("Percentage of deaths in each district during 3rd Wave")
```

```python
overall_deaths = pd.DataFrame()
overall_deaths['district'] = ""
overall_deaths['district'] = third_district_cases_mean.index.values
overall_deaths.set_index(['district'],inplace=True)
overall_deaths['first_deaths']=""
overall_deaths['first_deaths']= first_district_cases_mean['per_of_deaths'].copy()
overall_deaths['second_deaths']=""
overall_deaths['second_deaths']= second_district_cases_mean['per_of_deaths'].copy()
overall_deaths['third_deaths']=""
overall_deaths['third_deaths']= third_district_cases_mean['per_of_deaths'].copy()

overall_deaths.dropna(inplace=True)
overall_deaths
```

```python

overall_deaths[["first_deaths","second_deaths","third_deaths"]].plot(kind="barh",title="death rate",figsize=(40,120))
```
