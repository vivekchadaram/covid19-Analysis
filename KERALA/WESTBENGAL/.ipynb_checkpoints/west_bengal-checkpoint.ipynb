```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
# plt.figure(figsize=(30,12))
plt.rcParams['figure.figsize'] = [30,12]
```

```python
def get_month(date):
    date=date.split("-")
    if date[1]=="01":
        return "January"+date[2]
    if date[1]=="02":
        return "February"+date[2]
    if date[1]=="03":
        return "March"+date[2]
    if date[1]=="04":
        return "April"+date[2]
    if date[1]=="05":
        return "May"+date[2]
    if date[1]=="06":
        return "June"+date[2]
    if date[1]=="07":
        return "July"+date[2]
    if date[1]=="08":
        return "August"+date[2]
    if date[1]=="09":
        return "September"+date[2]
    if date[1]=="10":
        return "October"+date[2]
    if date[1]=="11":
        return "November"+date[2]
    if date[1]=="12":
        return "December"+date[2]
    
    
def get_year(date):
    date=date.split("-")
    return date[2]
def Histogram(x,y,xlabel,ylabel,title):
    x.plot(kind="hist",
        alpha=0.7,
        #         bins=[1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,15000],
        title=title,
        figsize=(12,8),
        color=["red"],rwidth=0.8
       )
    y.plot(kind="hist",
        alpha=0.7,
        #         bins=[1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,15000],
        title=title,
        figsize=(12,8),
        color=["blue"],rwidth=0.8
       )
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
def bar_graph(categories,cat_val,xlabel,ylabel,title=""):
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.bar(categories,cat_val,width=0.8)
    plt.title(title)
    plt.show()
# This is the code of linear_regression model also.........           
def linear_correlation(x,y):
    meanx=sum(x.values)/len(x)
    meany=sum(y.values)/len(y)
    if len(x.values)==len(y.values):
        num,den=0,0
        #numerator calculation
        a=0
        for i in range(len(x)):
            a+=x.values[i]*y.values[i]
            #         print(a)
        num=a*len(x)-(sum(x.values)*sum(y.values))
    
        #denominator calculation
        ax,ay=0,0
        for i in range(len(x)):
                ax+=x.values[i]**2
                ay+=y.values[i]**2 
        p1=(len(x)*ax-(sum(x.values)**2))**0.5
        p2=(len(y)*ay-(sum(y.values)**2))**0.5
        den=p1*p2
        #print(num,den)
        return num/den
    else:
        raise Exception("Invalid Dimension")
import datetime
import calendar
 
def get_day(date):
    day, month, year = (int(i) for i in date.split('-'))   
    dayNumber = calendar.weekday(year, month, day)
    days =["Monday", "Tuesday", "Wednesday", "Thursday",
                         "Friday", "Saturday", "Sunday"]
    return (days[dayNumber])

def set_index(x):
    if x=="May2020":
        return -3
    if x=="June2020":
        return -2
    if x=="July2020":
        return -1
    if x=="August2020":
        return 0
    if x=="September2020":
        return 1
    if x=="October2020":
        return 2
    if x=="November2020":
        return 3
    if x=="December2020":
        return 4
    if x=="January2021":
        return 5
    if x=="February2021":
        return 6
    if x=="March2021":
        return 7
    if x=="April2021":
        return 8
    if x=="May2021":
        return 9
    if x=="June2021":
        return 10
    if x=="July2021":
        return 11
    if x=="August2021":
        return 12
    if x=="September2021":
        return 13
    if x=="October2021":
        return 14
    if x=="November2021":
        return 15
    if x=="December2021":
        return 16
    if x=="January2022":
        return 17
def piechart(categories,cat_val,title=""):
    labels=categories
    values=cat_val
    #     explode = (0,0.1)
    plt.pie(values,labels=values,counterclock=False,autopct="%.2f", shadow=True)
    plt.title(title)
    plt.legend(labels,loc=3)
    plt.rcParams['figure.figsize'] = [10,50]
    plt.show()
def rearrange_date(x):
    arr=x.split("-")
    s="-"
    s1=s.join(arr[::-1])
    return s1
# vacc_month_sum[["vax_total_24h","vax_first_dose_24h","vax_sec_dose_24h"]].plot(kind="bar")
def plot_bar_graphs(data,values,xlabel="",ylabel="",title=""):
    data[values].plot(kind="bar",title="",xlabel=xlabel,ylabel=ylabel)
    
def get_date_month_year(case_info):
    case_info["day"]=case_info["date"].copy()
    case_info["month"]=case_info["date"].copy()
    case_info["year"]=case_info["date"].copy()
    case_info["day"]=case_info["day"].apply(lambda x:get_day(x))
    case_info["month"]=case_info["month"].apply(lambda x:get_month(x))
    case_info["year"]=case_info["year"].apply(lambda x:get_year(x))
    return case_info

def order_by_date(date):
    date=date.split("-")
    s=""
    for i in range(len(date)-1,-1,-1):
        s=s+date[i]

    return int(s)
     
```

# SUMMARY OF PART-1:

**PHASE-1:May-2020 to January-2021**
1. First wave in West Bengal had its impact till December-2020, we can see from the graph(**Graph1, Graph2, Graph3, Graph4,Graph9 and Graph10**) that there was a spike in number of positive cases, active cases per day,fatility.
2. Starting from June-2020, this wave continued till December-2020, making west bengal to be one-of the worst hit states after first-wave in the country. This small plateau which is visible from **June-2020 to December-2020** in the above mentioned graphs shows that it not only contribute to the number of positive case but also added to high fatility rate (can be seen in **Graph9 and Graph10**).
3. West-Bengal state govt had withdrawn state-wide lockdown on **28-August-2020,**(**source:** https://timesofindia.indiatimes.com/city/kolkata/west-bengal-govt-withdraws-state-wide-lockdown-on-august-28/articleshow/77511672.cms) was not the best decision as it didnt help to curb the virus spread. The active cases,death toll kept on increasing till December-2020 and the state's hospital and poor avalability of beds made the overall scenario worst.
4. Massive Durga-Puja Celebration from October-2020 shows and clearly defines the cause of surge in positive cases during the period **August-2020 to December-2020.** (**source:** https://theprint.in/opinion/newsmaker-of-the-week/bengal-durga-puja-2020-is-a-petri-dish-of-politics-and-pandemic/529856/), the state government spent nearly Rs 200 crore to provide financial assistance to 37,000 puja committees. CM also subsidised electricity bills and other license (including civic and fire) costs for the pandals. 
4. The state has very poor health infrastructure where it is seen from data that during **July-2020 to August-2020** the hospital and population ratio had reached nearly 1:1000000.
5. It is noted that Testing has direct role to determine number of active cases/total positive cases and hence the fatility rate. Testing comes before new cases, the more testing is done-> the possibility of positive cases increases. Testing graph (i.e **Graph13**) shows similar charateristics to that of positve cases graph (i.e **Graph2 and Graph6**) and death cases graph (i.e **Graph10**). And it is seen that people generally go for testing whenever they develop any symptoms, so it is quite clear from the data that due to massive festival celebration many people developed symptoms and tested positive, which justifies the elongated first wave in West Bengal till **December-2021.**


**PHASE-2: Feb-2021 to January-2022**
1. From December-2020 to January-2021 the positive cases decreased and thus the fatility rate.
2. Election Rallies started very comprehensively from Mid January-2020 for West Bengal election, this came out to be massive spread of the virus and the state was hit by full force during **March-2021 to June-2021** in the second-wave, the result was same, the recovery rate reduced during this period, there was increase in positive rate and fatiltiy rate.
3. Election rallies was one of the major means to spread the virus and thus the sceond wave.
4. There is again rise in case that can be seen from **Graph2 and Graph6** in **December 2021 to January-2022** marking the start of third wave in the state. Mass gathering and not following covid protocols during festival celebrations is again one of the major causes for this spike in cases.
5. Hospitilisation and fatility rate remains constant shows the death toll due to third wave could be relatively low.
6. **It is predicted that the number of positive cases in the state will maximum go as high as 1.5 to 2 times the cases witnessed during the second wave but significantly lower hospitalisation.**
